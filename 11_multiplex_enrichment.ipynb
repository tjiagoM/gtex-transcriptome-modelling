```python
import gseapy as gp
import pandas as pd
```

# Information on Multiplex Analysis 

In this notebook you will find code to check the enrichment of each group of genes identified in the multiplex analysis part.

##  Main function

This function will be run for each group:

```python
def print_lists_to_enrich(lists_to_enrich):
    # Making the column width big enough so it's possible to see the enriched terms.
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.width', 5000)

    for ind, list_enr in enumerate(lists_to_enrich):
        enr = gp.enrichr(gene_list=list_enr,
                                 organism='human',
                                 description="multiplex_all_tissues_" + str(ind),
                                 gene_sets='Reactome_2016',
                                 cutoff=0.05,
                                 no_plot=True, # skip plotting, because we don't need these figures
                                 outdir='results/EnrichClass')
        filtered_df = enr.results[enr.results['Adjusted P-value'] < 0.05].loc[:, ['Adjusted P-value', 'Term']]
        
        print(f'{" - ".join(list_enr)}:')
        if filtered_df.shape[0] == 0:
            print('No enrichment below adjusted p-value $< 0.05$.')
        else:
            display(filtered_df)
        print('\n')
    pd.reset_option('display.width')
    pd.reset_option('display.max_colwidth')
```

### All tissues

```python
all_tissues_to_enrich= [
    ['PRM1', 'PRM2'],
    ['PWP2', 'C21orf33'],
    ['HBA1', 'HBA2']
]

print_lists_to_enrich(all_tissues_to_enrich)
```
